```python
import numpy as np 
import pandas as pd 
```

```python

movies = pd.read_csv("Movies/tmdb_5000_movies.csv")
movies.head()
```

```python
movies.shape
```

```python
credits= pd.read_csv("Movies/tmdb_5000_credits.csv")
credits.head()
```

```python
credits.shape
```

## Merge 




```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(2)
```

```python
# Select important features 


movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head()
```

## Handle Missing and Duplicate values



```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True )
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
#Currently genres, keywords, cast, crew columns me JSON-like strings hain
#(e.g. '[{"id": 28, "name": "Action"}]')

#Ab unhe Python list of strings me convert karna hai.
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
# Top 3 cast 
def convert_cast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(2)
```

```python
#Extract Director Name from Crew
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
# Convert Overview into List of Words
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(2)
```

```python
#Remove Spaces from Multi-Word Names

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(2)
```

```python
#  Combine All Text Features → tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python

new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
# Convert List → String (for vectorization)

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# NLTK ka use karke hum:

# Stopword removal – irrelevant words (like “the”, “in”, “of”) hata dete hain.

# Stemming – words ke root form me convert karte hain (“playing” → “play”).

# Tokenization – text ko words me todte hain safely.

```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        if i not in stopwords.words('english'):
            y.append(ps.stem(i))
    return " ".join(y)


```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
# Text Vectorization (Convert Text → Numbers)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Measure Similarity Between Movies

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
# Recommendation Function



def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
# Test the Model

recommend('Avatar')
```
